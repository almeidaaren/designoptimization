```python
#gradient descent
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
from numpy import gradient
def mean_squared_error(y_act, y_predicted): 
    # Calculating the loss or cost
    cost = np.sum((y_act-y_predicted)**2) / len(y_act)
    return cost
# Gradient Descent Function
# Here iterations, learning_rate, stopping_threshold
#assigning different values required for gradient descent
def gradient_descent(x1,x2,x3,y, iterations = 3000, learning_rate = 0.0001,stopping_threshold = 1e-3): 
    # Initializing values
    current_weight = 0
    current_bias = 0.1
    iterations = iterations
    learning_rate = learning_rate
    n = float(len(x1)) 
    costs = []
    weights = []
    previous_cost = None 
    # for iterations
    for i in range(iterations): 
        # predicting Y
        y_predicted = (current_weight * x1)+(2*current_weight * x2)+(3*current_weight * x3)-1+ current_bias
        # Calculating the current cost
        current_cost = mean_squared_error(y, y_predicted)
        # If the change in cost is less than or equal to
        # stopping_threshold and stop gradient descent
        if previous_cost and abs(previous_cost-current_cost)<=stopping_threshold:
            break
         
        previous_cost = current_cost
        costs.append(current_cost)
        weights.append(current_weight)
        # Calculating the gradients
        weight_derivative = -(2/n) * sum(x1 * (y-y_predicted) + 2*x2 * (y-y_predicted) + 3*x3 * (y-y_predicted)-1)
        bias_derivative = -(2/n) * sum(y-y_predicted)
         
        # Updating weights and bias
        current_weight = current_weight - (learning_rate * weight_derivative)
        current_bias = current_bias - (learning_rate * bias_derivative)
                 
        # Printing the parameters for each 1000th iteration
        print(f"Iteration {i+1}: Cost {current_cost}, Weight \ {current_weight}, Bias {current_bias}")
        # ploting the weights and cost for all iterations
    plt.figure(figsize = (10,9))
    plt.plot(weights, costs)
    plt.scatter(weights, costs, marker='1', color='red')
    plt.title("Cost vs Weights")
    plt.ylabel("Cost")
    plt.xlabel("Weight")
    plt.show()
    return current_weight, current_bias
    # initial Data
def main():
    X1 = np.array([2.50234527, 3.42680403, 1.53035803, 7.47563963, 9.81320787,
       8.95588857, 4.68719623, 0.29732685, 4.61864377, 8.81681754])
    X2 = np.array([3.50234527, 5.42680403, 6.53035803, 4.47563963, 5.81320787,
       4.95588857, 4.68719623, 6.29732685, 4.61864377, 3.81681754])
    X3 = np.array([4.50234527, 5.42680403, 3.53035803, 2.47563963, 7.81320787,
       7.95588857, 5.68719623, 9.29732685, 5.61864377, 9.81681754])
    Y = np.array([1.70700585, 8.77759598, 2.5623823 , 1.54663223, 7.23092513,
       0.72360244, 2.89250373, 7.37989686, 4.84715332, 6.87721319])
# Estimating weight and bias using gradient descent
    estimated_weight, estimated_bias = gradient_descent(X1,X2,X3,Y, iterations=3000)
    print(f"Estimated Weight: {estimated_weight}\nEstimated Bias: {estimated_bias}")

# Making predictions using estimated parameters
    Y_pred = estimated_weight*X1 + estimated_weight*2*X2 + estimated_weight*3*X3+  estimated_bias-1

# Plotting the regression line
    plt.figure(figsize = (10,9))
    plt.scatter(Y_pred, Y, marker='o', color='red')
    plt.plot([min(X1), max(X1)], [min(Y_pred), max(Y_pred)], color='blue')
    plt.plot([min(X2), max(X2)], [min(Y_pred), max(Y_pred)], color='red')
    plt.plot([min(X3), max(X3)], [min(Y_pred), max(Y_pred)], color='green')
    plt.title("Y Predicted vs Y")
    plt.xlabel("Y Predicted")
    plt.ylabel("Y")
    plt.show()
    
if __name__=="__main__":
    main()
```
